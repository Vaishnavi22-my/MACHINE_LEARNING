```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
from sklearn.datasets import fetch_openml
titanic = fetch_openml('titanic' , version = 1 , as_frame = True)
```

```python
X = titanic.data
y = titanic.target
```

```python
X
```

```python
y
```

```python
df = pd.concat([X , y] , axis =1)
```

```python
df
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df.drop(['cabin' , 'boat' , 'home.dest'] , axis = 1 , inplace = True)
```

```python
from sklearn.impute import SimpleImputer
print(f'Number of missing values before imputation {df.age.isnull().sum()}')
si = SimpleImputer(strategy = 'mean')
df['age'] = si.fit_transform(df[['age']])
print(f'Number of missing values after imputation {df.age.isnull().sum()}')
```

```python
df['fare'] = si.fit_transform(df[['fare']])
si1 = SimpleImputer(strategy = 'most_frequent')
df['embarked'] = si1.fit_transform(df[['embarked']]).ravel()
```

```python
df.drop(['body'] , axis = 1, inplace = True)
```

```python
df.isnull().sum()
```

```python
df
```

```python
df.drop(['name' , 'ticket'] , axis = 1 , inplace = True)
```

```python
df
```

```python
threshold = 3
from scipy.stats import zscore

numerical_features = ['fare' , 'age']
z_scores = np.abs(zscore(df[numerical_features]))
print(z_scores)
```

```python
outliers = np.where(z_scores >threshold)
print(outliers)
```

```python
from scipy.stats.mstats import winsorize

df['age'] =  winsorize(df['age'] , limits = [0.15 , 0.15])
df['fare'] =  winsorize(df['fare'] , limits = [0.15 , 0.15])
```

```python
threshold = 3
from scipy.stats import zscore

numerical_features = ['fare' , 'age']
z_scores = np.abs(zscore(df[numerical_features]))
print(z_scores)
```

```python
outliers = np.where(z_scores >threshold)
print(outliers)
```

```python
from sklearn.preprocessing import OrdinalEncoder
oe = OrdinalEncoder()
df['sex'] = oe.fit_transform(df[['sex']])
```

```python
df
```

```python
df['embarked'] = oe.fit_transform(df[['embarked']])
```

```python
df
```

```python
X = df.iloc[: , :-1]
y = df.iloc[: , -1]
```

```python
from sklearn.model_selection import train_test_split
X_train , X_test , y_train , y_test = train_test_split(X , y , test_size = 0.3 , random_state = 42)
```

```python
from sklearn.ensemble import AdaBoostClassifier

ad = AdaBoostClassifier(n_estimators = 100 , learning_rate = 1 , random_state = 0 )
ad.fit(X_train , y_train)
```

```python
y_pred = ad.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_pred , y_test)
```

HyperTuning

```python
from yellowbrick.model_selection import validation_curve

num_est = [100 , 250 , 200 , 50 , 150 , 75 ]
print(validation_curve(AdaBoostClassifier() , X= X_train , y=y_train ,param_name = 'n_estimators' , param_range = num_est , scoring = 'accuracy' , cv = 3))
```

```python
from yellowbrick.model_selection import validation_curve

lr_vals = [0.1 ,0.01 , 0.2 , 0.5 , 0.8 , 1]
print(validation_curve(AdaBoostClassifier() , X= X_train , y=y_train ,param_name = 'learning_rate' , param_range = lr_vals , scoring = 'accuracy' , cv = 3))
```

```python
ad1 = AdaBoostClassifier(n_estimators = 200 , learning_rate = 0.5 , random_state = 0 )
ad1.fit(X_train , y_train)

pred2 = ad1.predict(X_test)
```

```python
print(accuracy_score(pred2 , y_test))
```

```python
feature_scores = pd.Series(ad1.feature_importances_ , index = X_train.columns).sort_values(ascending = False)
print(feature_scores)
```

```python
sns.barplot(x = feature_scores , y = feature_scores.index)
plt.xlabel('Feature Importance Score')
plt.ylabel('Features')
plt.title('Visualizing Important Features')
plt.show()
```
